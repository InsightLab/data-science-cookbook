```python
total,_ = df.shape
for atributo in df.columns[:-1]:
    for classificador in df["classes"].unique():
        for valoresDistintosColuna in df[atributo].unique():
            x ,_ =df[ (df[atributo] == valoresDistintosColuna) & (df["classes"] == classificador) ].shape
            y ,_ =df[ df[atributo] == valoresDistintosColuna ].shape
            dicionario[atributo,valoresDistintosColuna] = y/total
            dicionario[atributo,valoresDistintosColuna,classificador] = x / total
        
        quantidadeTotalClasse,_ = df[df["classes"]==classificador].shape
        dicionario[classificador] = quantidadeTotalClasse/total
```
