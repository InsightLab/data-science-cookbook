#### Lendo arquivo de dados:

```python
import csv
 
def loadCsv(filename):
    lines = csv.reader(open(filename, "r"))
    dataset = list(lines)
    for i in range(len(dataset)):
        dataset[i] = [x for x in dataset[i]]
    return dataset
```

#### Transformando dados categóricos em dados numéricos:

```python
def encode(dataset):
    encoded_copy = dataset
    feature_list = []
    #Obtendo numero de colunas do dataset
    for i in range(len(dataset[0])):
        #Para cada coluna criamos uma lista, que irá armazenar as features distintas
        listE = []
        feature_list.append(listE)
        
    #Agora verificamos o número distinto de features em cada coluna e inserimos na nossa lista de features
    for i in range(len(dataset[0])):
        for j in dataset:
            if(j[i] not in feature_list[i]):
                feature_list[i].append(j[i])
                
    #Agora percorreremos a cópia do dataset e compararemos com nossa lista de features. Caso um elemento 
    #for igual à feature, o substituiremos pelo seu índice na lista de features.
    
    for i in range(len(encoded_copy[0])):
        for j in encoded_copy:
            if(j[i] in feature_list[i]):
                j[i] = feature_list[i].index(j[i])
            
    return encoded_copy
```

#### Dividindo conjunto em treino e teste

```python
import random
def splitDataset(dataset, splitRatio):
    trainSize = int(len(dataset) * splitRatio)
    trainSet = []
    copy = dataset
    while len(trainSet) < trainSize:
        index = random.randrange(len(copy))
        trainSet.append(copy.pop(index))
    return [trainSet, copy]
```

```python
data_split = splitDataset(data, 0.67)
```

```python
train_set = data_split[0]
```

```python
test_set = data_split[1]
```

### Sumarizando os dados

#### Separando os dados por classe

```python
def separateByClass(dataset):
    separated = {}
    for i in range(len(dataset)):
        vector = dataset[i]
        if (vector[-1] not in separated):
            separated[vector[-1]] = []
        separated[vector[-1]].append(vector)
    return separated
```

#### Calculando média e desvio padrão

```python
import math
def mean(numbers):
    return sum(numbers)/float(len(numbers))
 
def stdev(numbers):
    avg = mean(numbers)
    variance = sum([pow(x-avg,2) for x in numbers])/float(len(numbers)-1)
    return math.sqrt(variance)
```

#### Sumarizando os dados

```python
def summarize(dataset):
    summaries = [(mean(attribute), stdev(attribute)) for attribute in zip(*dataset)]
    del summaries[-1]
    return summaries
```

#### Sumarizando atributos por classe

```python
def summarizeByClass(dataset):
    separated = separateByClass(dataset)
    summaries = {}
    for classValue, instances in separated.items():
        summaries[classValue] = summarize(instances)
    return summaries
```

### Realizando a predição

#### Calcular a função de densidade da probabilidade gaussiana

```python
def calculateProbability(x, mean, stdev):
    if(stdev == 0):
        stdev = 10000
    exponent = math.exp(-(math.pow(x-mean,2)/(2*math.pow(stdev,2))))
    return (1 / (math.sqrt(2*math.pi) * math.pow(stdev, 2))) * exponent
```

#### Calcular probabilidades das classes

```python
def calculateClassProbabilities(summaries, inputVector):
    probabilities = {}
    for classValue, classSummaries in summaries.items():
        probabilities[classValue] = 1
        for i in range(len(classSummaries)):
            mean, stdev = classSummaries[i]
            x = inputVector[i]
            probabilities[classValue] *= calculateProbability(x, mean, stdev)
    return probabilities
```

#### Fazer a predição

```python
def predict(summaries, inputVector):
    probabilities = calculateClassProbabilities(summaries, inputVector)
    bestLabel, bestProb = None, -1
    for classValue, probability in probabilities.items():
        if bestLabel is None or probability > bestProb:
            bestProb = probability
            bestLabel = classValue
    return bestLabel
```

#### Fazer várias predições

```python
def getPredictions(summaries, testSet):
    predictions = []
    for i in range(len(testSet)):
        result = predict(summaries, testSet[i])
        predictions.append(result)
    return predictions
```

#### Calcular acurácia

```python
def getAccuracy(testSet, predictions):
    correct = 0
    for i in range(len(testSet)):
        if testSet[i][-1] == predictions[i]:
            correct += 1
    return (correct/float(len(testSet))) * 100.0
```

```python
summarized_data = summarizeByClass(train_set)
```

```python
predictions = getPredictions(summarized_data, test_set)
```
